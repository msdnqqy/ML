# 聚类算法

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans,KMeans
from sklearn import metrics
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
#生成数据
x,y=make_blobs(n_samples=1000,n_features=2,centers=[[-1,-1],[0,0],[1,1],[2,2]],cluster_std=[0.4,0.2,0.2,0.1],random_state=0)
```

```python
plt.plot(x[:,0],x[:,1],'go')
```

```python
#应用kmeans进行聚类
for i in np.arange(2,6):
    plt.subplot(2,2,i-1)
    y_pre=MiniBatchKMeans(n_clusters=i,batch_size=100,random_state=0).fit_predict(x)
    score=metrics.calinski_harabaz_score(x,y_pre)
    plt.scatter(x[:,0],x[:,1],c=y_pre)
    plt.text(.99,.01,('k=%d,score:%.2f'%(i,score)))
    
plt.show()
```
