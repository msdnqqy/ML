## 1.sklearn 实现随机森林

### 1.1集成学习不是一个单独的机器学习算法，而是通过在数据上构建多个模型，集成所有模型的建模结果。集成算法会考虑多个评估器的建模结果，汇总之后得到一个综合的结果，以此来获取比单个模型更好的回归或分类表现，得到比单一模型更好的泛化性能和准确率

#### 1.1多个模型集成的模型叫做集成评估器，组成集成评估器的每个模型叫基学习器。通常来说有三种集成算法：装袋法（bagging），提升法（boosting），stacking

##### 1.1.1装袋法核心思想是构建多个相互独立的评估器。具体做法是有放回抽样，拿到T个不同的训练集合。从而生成T个不同的基学习器。再将这些基学习器进行结合（如简单投票法，简单平均法）

##### 1.1.2提升法 boosting，将弱学习器提升为强学习器的方法算法。具体做法是先从初始训练集生成一个基学习器，再根据基学习器的表现对训练集的样本分布进行调整，增加被分类错误的样本在下一次训练集中的出现概率或给该样本给予更高的权重，然后基于调整后的训练集训练下一个分类器，如此反复进行，直至基学习器达到事先指定的值T，最终将这T个学习器进行加权集合

#### 1.2sklearn中的集成算法都在包ensemble中


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
wine
```

```python
wine.data,wine.target
```

```python
from sklearn.model_selection import train_test_split
```

### 一.随机森林的基础实现

```python
x_train,x_test,y_train,y_test=train_test_split(wine.data,wine.target,test_size=0.3)
```

```python
clf=DecisionTreeClassifier(random_state=0)
rfc=RandomForestClassifier(random_state=0)

clf.fit(x_train,y_train)
rfc.fit(x_train,y_train)

score_c=clf.score(x_test,y_test)
score_r=rfc.score(x_test,y_test)

print("score_c:",score_c,"score_r",score_r)
```

#### 通常来说随机森林的效果优于单一的决策树。

### 二.随机森林交叉验证效果对比

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

#创建决策树模型和随机森林模型
random_model=RandomForestClassifier(n_estimators=25)
decision_model=DecisionTreeClassifier()

random_model_score=cross_val_score(random_model,wine.data,wine.target,cv=10)#10折交叉验证
decision_model_score=cross_val_score(decision_model,wine.data,wine.target,cv=10)

plt.plot(range(1,11),random_model_score,label="random_model_score")
plt.plot(range(1,11),decision_model_score,label="decision_model_score")
plt.legend()
plt.show()
```

### 三.决策树&随机森林十组交叉验证

```python
random_model_scores=[]
decision_model_scores=[]
for i in range(10):
    random_model=RandomForestClassifier(n_estimators=25)
#     random_model.fit(x_train,y_train)
    random_model_score=cross_val_score(random_model,wine.data,wine.target,cv=10).mean()
    random_model_scores.append(random_model_score)
    
    decision_model=DecisionTreeClassifier()
    decision_model_score=cross_val_score(decision_model,wine.data,wine.target,cv=10).mean()
    decision_model_scores.append(decision_model_score)
    

    
plt.plot(range(1,11),random_model_scores,label="RandomForestClassifier")
plt.plot(range(1,11),decision_model_scores,label="DecisionTreeClassifier")
plt.legend()
plt.show()
    
```

###### 可以看到随机森林是远优于单个决策树模型的

### 四.n_estimators学习曲线（调参）

```python
superpa=[]
for i in range(100):
    if i%20==0:
        print("step:",i)
    random_model=RandomForestClassifier(n_estimators=i+1,n_jobs=-1)
    random_model_score=cross_val_score(random_model,wine.data,wine.target,cv=10).mean()
    superpa.append(random_model_score)
    
print(max(superpa),superpa.index(max(superpa)))
plt.figure(figsize=[20,5])
plt.plot(range(1,101),superpa)
plt.show()
```

### 七.调参实例--sklearn 乳腺癌数据

```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#导入数据
cancer=load_breast_cancer()
cancer.data.shape,cancer.target.shape,cancer.data[0:5],cancer.target[0:5]
```

```python
#创建简单的随机森林进行验证
random_model=RandomForestClassifier(n_estimators=100,random_state=0)
random_model_score=cross_val_score(random_model,cancer.data,cancer.target,cv=10).mean()
random_model_score
```

### 八.使用学习曲线来对n_estimators进行调参

```python
scores=[]
for i in range(0,200,10):
    random_model=RandomForestClassifier(n_estimators=i+1
                                        ,n_jobs=-1
                                       ,random_state=0)
    random_model_score=cross_val_score(random_model,cancer.data,cancer.target,cv=10).mean()
    scores.append(random_model_score)
    
#打印最大score及其位置
print(max(scores),scores.index(max(scores)))
plt.figure(figsize=[20,5])
plt.plot(range(1,201,10),scores)
plt.show()
```

### 九.网格搜索

```python
param={"max_depth":np.arange(1,20,1)}

random_model=RandomForestClassifier(n_estimators=115
                                   ,random_state=0
                                   )
GS=GridSearchCV(random_model,param,cv=10)
GS.fit(cancer.data,cancer.target)
GS.best_params_,GS.best_score_
```
